```python
df23 = df23.drop('RC_UNITS', axis=1)
```

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na=[features for features in df23.columns if df23[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values
m1=[]
for feature in features_with_na:
    print(feature, np.round(df23[feature].isnull().mean(), 4),  ' % missing values\n')
    percentage_null = df23[feature].isnull().sum() / len(df23) * 100
    print(percentage_null,'\n\n')
    if percentage_null>50:
        m1.append(feature)
```

```python
colours=['r','b','g','y','m','c','y','k','w']
for feature in features_with_na:
    data = df23.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['TOTAL_VALUE'].median().plot.bar(color=colours)
    plt.title(feature)
    plt.show()
```

# Discrete Variable

```python
## Numerical variables are usually of 2 type
## 1. Continous variable and Discrete Variables

discrete_feature=[feature for feature in numerical_features if len(df23[feature].unique())<50 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
discrete_feature
```

```python
df23[discrete_feature]
```

```python
## Lets Find the realtionship between them and Sale PRice
colours=['r','b','g','y','m','c','y','k','w']

for feature in discrete_feature:
    data=df23.copy()
    data.groupby(feature)['TOTAL_VALUE'].median().plot.bar(color=colours)
    plt.xlabel(feature)
    plt.ylabel('TOTAL_VALUE')
    plt.title(feature)
    plt.show()
```

# Continuous Variable

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print("Continuous feature Count {}".format(len(continuous_feature)))
continuous_feature
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_feature:
    data=df23.copy()
    data[feature].hist(bins=25,color='g')
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
## We will be using logarithmic transformation


for feature in continuous_feature:
        data=df23.copy()
#     if 0 in data[feature].unique():
#         pass
#     else:
        data[feature]=np.log(data[feature])
        data['TOTAL_VALUE']=np.log(data['TOTAL_VALUE'])
        plt.scatter(data[feature],data['TOTAL_VALUE'])
        plt.xlabel(feature)
        plt.ylabel('TOTAL_VALUE')
        plt.title(feature)
        plt.show()
```

```python
for feature in continuous_feature:
        data=df23.copy()
#     if 0 in data[feature].unique():
#         pass
#     else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
