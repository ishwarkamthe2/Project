# Missing Values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in df23.columns if df23[feature].isnull().sum()>1 and df23[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df23[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(df23,features_nan):
    data=df23.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

df23=replace_cat_feature(df23,features_nan)

df23[features_nan].isnull().sum()
```

```python
df23.head()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in df23.columns if df23[feature].isnull().sum()>1 and df23[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values
cc=0
for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df23[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=df23[feature].median()
    
    ## create a new feature to capture nan values
#     df23[feature+'nan']=np.where(df23[feature].isnull(),1,0)
    df23[feature].fillna(median_value,inplace=True)
    
df23[numerical_with_nan].isnull().sum()
```

# Handling Rare Categorical Feature

```python
# We will remove categorical variables that are present less than 1% of the observations
```

```python
categorical_features=[feature for feature in df23.columns if df23[feature].dtype=='O']

# for feature in categorical_features:
#     print('The feature is {} and number of categories are {}'.format(feature,len(df23[feature].unique())))

# x=categorical_features
#x = list(filter(lambda x: x not in ['ST_NAME','UNIT_NUM','OWNER','OWNER MAIL ADDRESS'], x))
```

```python
for feature in categorical_features:
    temp=df23.groupby(feature)['TOTAL_VALUE'].count()/len(df23)
    temp_df=temp[temp>0.01].index
    df23[feature]=np.where(df23[feature].isin(temp_df),df23[feature],'Rare_var')
```
